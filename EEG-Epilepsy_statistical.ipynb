```python
# Importing Libraries
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load EEG data from CSV
file_path = 'Seizure-EEG.csv'  # Replace with the path to your CSV file
data = pd.read_csv(file_path, header=None)
```

```python
# Calculate statistical parameters
stats = data.describe().T  # Basic statistics
stats['variance'] = data.var()  # Variance
stats['skewness'] = data.skew()  # Skewness
stats['kurtosis'] = data.kurtosis()  # Kurtosis
```

```python
# Save statistical parameters to a new CSV file
stats.to_csv('eeg_Epilepsy_statistical_parameters.csv')
print("Statistical parameters saved to 'eeg_statistical_parameters.csv'")
```

```python
# Visualize the EEG data and statistical parameters
plt.figure(figsize=(16, 12))
```

```python
# Plot 1: Time series of each EEG channel
for i in range(data.shape[1]):
    plt.plot(data.index, data.iloc[:, i], label=f'Channel {i+1}')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('EEG Channel Time Series')
plt.legend()
plt.show()
```

```python
# Plot 2: Heatmap of correlation between EEG channels
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of EEG Channels")
plt.show()
```

```python
# Plot 3: Box plot for each EEG channel
plt.figure(figsize=(12, 8))
sns.boxplot(data=data)
plt.title("Box Plot of EEG Channels")
plt.xlabel("Channels")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Plot 4: Histogram for each EEG channel
data.hist(figsize=(16, 12), bins=20)
plt.suptitle("Histogram of Each EEG Channel")
plt.show()
```

```python
# Plot 5: KDE plot for each EEG channel
plt.figure(figsize=(12, 8))
for i in range(data.shape[1]):
    sns.kdeplot(data.iloc[:, i], label=f'Channel {i+1}')
plt.title("KDE Plot of EEG Channels")
plt.xlabel("Amplitude")
plt.legend()
plt.show()
```
